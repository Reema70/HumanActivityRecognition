# #How accelerometer data looks visually:


```python
def plot_activity(activity, data):
    extractRow = (data['activity'] == activity)
    data1 = data[ extractRow]
    data1 = data[['x','y','z']]
    data1 = data[:2000]
    
    ax = data1.plot(subplots=True, figsize=(16, 12), title= activity)
```

```python
for activity in activities:
    plot_activity( activity,data)
```
